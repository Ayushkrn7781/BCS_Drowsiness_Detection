```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Model
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from tensorflow.keras import layers

```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()
x_test[59]
```

```python
#normalization of the data
x_train=x_train/255
x_test=x_test/255
```

```python
# y_train = keras.utils.to_categorical(y_train)
# y_test=keras.utils.to_categorical(y_test)
# y_train.shape
```

```python
input_layer=layers.Input(shape=(28,28,1))
image=layers.Conv2D(filters=32, kernel_size=(3, 3),strides=(2,2) ,activation='relu')(input_layer)
#image.shape [13, 13, 32] 32 layers of grid 26*26 output=(28-3/stride=2)+1
image.shape
image=layers.MaxPooling2D(pool_size=(2,2),strides=1)(image)
#image.shape [12,12,32] output=(26-2/stride=1)+1

image=layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu')(image)
image=layers.MaxPooling2D(pool_size=(2,2),strides=1)(image)
image.shape
image=layers.Conv2D(filters=128, kernel_size=(3, 3), activation='relu')(image)
image=layers.MaxPooling2D(pool_size=(2,2),strides=1)(image)

#image.shape after the three layer [6,6,128] have 128 layers of 6*6
```

```python
print(image)
```

```python
image=layers.Flatten()(image) # array of 1 dimension of length 6*6*128=4,608
#we are doint it because dense function read in 1d array
image= layers.Dense(units=128, activation='relu')(image)
#now array contain 128 elements as
# Dense layer take input multiply with thier wieghts and add bias
image= layers.Dense(units=64, activation='relu')(image)
image= layers.Dense(units=10, activation='softmax')(image)
#initially we make are out as a class
#softmax function=a function that turns a vector of K real values into a vector of K real values that sum to 1 softmax funtion
#so it return the array having 10 entries.
```

```python
cnn_model=Model(inputs=input_layer, outputs=image)
#it compacting all the funtions and algos into a single model After this model is ready to work for prediction
```

```python
cnn_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
#it tells fit to how to optimize the model and update weights and bias
#loss function tell that how much model predictions is far from actual data

#To monitor accuracy during training we use matrices['accuracy']
```

```python
print(x_train.shape)
print(y_train.shape)
```

fit funtion compare the output by x_train and y_train

```python
cnn_model.fit(x_train, y_train, epochs=5)
#compare the predicted output and anctual output and using the adam optimizer increase accuracy epoch by epoch
```

```python
loss,accuracy=cnn_model.evaluate(x_test,y_test)

#Now checking accuracy and loss funtion for our test data set using evaluate funtion
```

```python
test=x_test[59]
test= np.expand_dims(test, axis=0)
#it creating one image having 28*28 pixel having grayecale values
test.shape
```

```python
ans=cnn_model.predict(test)
#it returns the array having 10 entries having only one entry is 1 and rest all are zero
#which is our predicted output index
predicted_output=np.argmax(ans)
predicted_output
#.argmax -> return index of max element
#which is our predicted output
```
